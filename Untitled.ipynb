```python
from corner import corner
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as ss
import seaborn as sns
```

```python
df = pd.read_csv('World Happiness Report 2019.csv')
df.head()
```

```python
def plot_relational_plot(df):
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.scatter(df['GDP per capita'], df['Healthy life expectancy'],
                marker='o', s=50, c=df['Score'], cmap='RdYlGn', edgecolor='black')
    plt.colorbar(label='Score')


    plt.xlabel('GDP per capita')
    plt.ylabel('Healthy life expectancy')
    plt.title('GDP per capita vs Healthy life expectancy')

    plt.show()
    plt.savefig('relational_plot.png')

    return
```

```python
plot_relational_plot(df)
```

```python
def plot_categorical_plot(df):
    fig, ax = plt.subplots(figsize=(6, 30))
    ax.barh(df['Country or region'], df['Score'], color='blue')  
    
    ax.set_xlabel('Score')
    ax.set_ylabel('Country or Region')
    ax.set_title('Categorical Plot of Countries and Scores')
    
    plt.tight_layout()
    plt.savefig('categorical_plot.png')
    plt.show()
```

```python
plot_categorical_plot(df)
```

```python
def plot_statistical_plot(df, columns, method):
    
    fig, ax = plt.subplots(dpi=144)
    corr_matrix = df[columns].corr(method=method)

    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    
    sns.heatmap(corr_matrix, 
                ax=ax, 
                vmin=-1, vmax=1, 
                cmap='RdBu', 
                annot=True, 
                mask=mask, 
                linewidths=0.5, 
                fmt='.2f')

    plt.title(f'{method.capitalize()} Correlation Heatmap')
    plt.savefig('statistical_plot.png')
    return
```

```python
plot_statistical_plot(df, columns, method='pearson')
plot_statistical_plot(df, columns, method='kendall')
columns = ['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices' ]
```
